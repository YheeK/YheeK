I will start testing ANOVA, before testing linear regression. 

```python
fit_sort = ols(formula='views ~ C(sort)', data = Ted_Nominal).fit()
anova_lm(fit_sort)
```

Sort is a nominal factor of exogenous category. As P-value is lower than 0.05, this result is statistically trusted. In addition, the F-value is 13.37 which is bigger than critical value(3.06) 

```python
Ted_Nominal['intercept'] = 1
df_sort = Ted_Nominal.join(pd.get_dummies(Ted_Nominal.sort))
sm.OLS(df_sort['views'], df_sort[['intercept', 'ted', 'tedx', 'tedwomen']]).fit().summary()
```

When the result appears statistically significant, I tested in linear regression as dummy variable

Y =  ( 1.37e+06)+(1.483e+06)X1+(-1.231e+05)X2+(9616.4695)X3. 

```python
Ted_Nominal['intercept'] = 1
df_topic = Ted_Nominal.join(pd.get_dummies(Ted_Nominal.topic))
sm.OLS(df_topic['views'], df_topic[['intercept', 'entertainment','science','personalgrowth','technology', 'design','business','health']]).fit().summary()
```

You can actually find the ANOVA test in linear regression table.

```python
Ted_Nominal.speed_change.dropna(axis=0, inplace=True)
cnt,bins=np.histogram(Ted_Nominal.speed_change,bins=2)
print(cnt)
print(bins)
```

```python
Ted_Nominal['sc_bins'] = pd.cut(x=Ted_Nominal.speed_change, bins= bins,
                       labels = ['No', 'Yes'],
                       include_lowest = True)
Ted_Nominal.sc_bins
```

I put Nominal factors in the dummy variable with the use of pandas, but here i already had data as 0 and 1, so i needed to lable them to show thier identitiy in the linear regression.

```python
Ted_Nominal['intercept'] = 1
df_Speed_C = Ted_Nominal.join(pd.get_dummies(Ted_Nominal.sc_bins))
sm.OLS(df_Speed_C['views'], df_Speed_C[['intercept', 'Yes','No']]).fit().summary()
```
